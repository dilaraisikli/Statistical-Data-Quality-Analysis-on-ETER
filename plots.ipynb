# Plots of result : for all institutions

### Total Number of Institution for each Country per Year

```python
fig = plt.figure(figsize=(14,5))
ax = fig.add_axes([0,0,1,1])
ax = sns.barplot(x="Country Code", y="Total number of institutions", hue="Reference year", data=result)
sns.set_palette('bright')
plt.title('Total Number of Institution for each Country per Year')
plt.xlabel('Country Code')
plt.ylabel('Total number of institution')
plt.show()
```

```python
fig = plt.figure(figsize=(14,5))
ax = fig.add_axes([0,0,1,1])
ax = sns.boxplot(x="Country Code", y='Total students enrolled ISCED 5-7', hue="Reference year", data=result);
sns.set_palette('bright')
plt.title('Total Number of Institution for each Country per Year')
plt.xlabel('Country Code')
plt.ylabel('Total number of institution')
plt.show()
```
