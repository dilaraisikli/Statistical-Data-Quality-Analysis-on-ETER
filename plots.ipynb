### 10 most total number of Total Current expenditure (EURO) with their total institutions number across Europe in 2016

Financial data in ETER deal with two distinct aggregates:<br/>
    1-Revenues, i.e. the amount of money received by the HEI, either as a general
allocation from the state/public authorities or other funding for research, education
and other services (which could be coming from public and/or private sources).<br/>
    2-Expenditures, i.e. the amount of money spent by an HEI in order to manage its
activities, for example for paying its employees, external services, general
expenditures and maintenance costs, as well as capital costs.

Current expenditure:<br/>
This variable includes all current expenditures of the considered HEI for its activities,
including salaries, payment for services, repairs and maintenance expenditures, electricity, financial costs, etc. . It excludes depreciations and capital costs even if they
are included in the operational account (cash accounting principle).
**This variable measures the total level of expenditure of higher education institutions in
the reference period.** The perimeter corresponds to expenditures inside institutions as
defined in the UOE manual (UOE manual section 2.9) covering all types of goods and
services, It includes both expenditures for educational core goods and services, R&D
expenditures and other non-instructional goods and services, like meals and
transportation, when directly made by HEI institutions.
Special case. Expenditure for hospitals. Consistently with the UOE manual, expenditure
by or on teaching hospitals (sometimes referred to as academic hospitals oruniversity
hospitals) is excluded from educational expenditure, particularly all costs of patient care
and other general expenses of academic hospitals, even if such expenses are paid by the
education authorities. However, expenditure by or on teaching hospital that it is directly
and specifically related to the training of medical students, expenditure on R&D at
teaching hospitals to the extent that it is included in the OECD/DSTI data collection on
R&D are included.
Remark: the sum of personnel, non-personnel and unclassified expenditure should equal
operating expenditure.
Variable: operating expenditure.
Type of variable: integer, national currency.
Data source: NSA.

```python
result[result['Reference year'] == 2016][['Country Code','Total number of institutions',
                                          'Total Current expenditure (EURO)']
                                        ].sort_values(by =['Total Current expenditure (EURO)',
                                                           'Total number of institutions'], ascending=False).head(10)
```

### Proportion of Students Enrolled ISCED 5-7 in 20 first institutions with most number of students across Europe in 2016

```python
top_20_st = result[result['Reference year'] == 2016][['Country Code','Total students enrolled at ISCED 5',
                                        'Total students enrolled at ISCED 6','Total students enrolled ISCED 7 long degree',
                                        'Total students enrolled at ISCED 7','Total students enrolled ISCED 5-7'
                                        ]].sort_values(by =['Total students enrolled ISCED 5-7'], ascending=False).head(20)
```

```python
sns.set(style="whitegrid")

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(15, 7))

sns.set_color_codes("pastel")
# sns.set_palette(sns.palplot(sns.cubehelix_palette(8)))
sns.barplot(x="Total students enrolled ISCED 5-7", y="Country Code", data=top_20_st,
            label="Total students enrolled ISCED 5-7", color="b")
sns.set_color_codes("bright")
sns.barplot(x="Total students enrolled at ISCED 5", y="Country Code", data=top_20_st,
            label="Total students enrolled at ISCED 5", color="purple")
sns.set_color_codes("muted")
sns.barplot(x="Total students enrolled at ISCED 6", y="Country Code", data=top_20_st,
            label="Total students enrolled at ISCED 6", color="b")
sns.set_color_codes("bright")
sns.barplot(x="Total students enrolled at ISCED 7", y="Country Code", data=top_20_st,
            label="Total students enrolled at ISCED 7", color="b")
sns.set_color_codes("dark")
sns.barplot(x="Total students enrolled ISCED 7 long degree", y="Country Code", data=top_20_st,
            label="Total students enrolled ISCED 7 long degree", color="b")

# Add a legend and informative axis label
ax.legend(ncol=2, loc="right", frameon=True)
ax.set(ylabel="Country Code",
       xlabel="Total number of students")
sns.despine(left=True, bottom=True)
```
